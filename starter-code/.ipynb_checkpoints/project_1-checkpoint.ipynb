## Step 1: Open the `sat_scores.csv` file. Investigate the data, and answer the questions below.


##### 1. What does the data describe?

The data describes the math and verbal SAT scores for each state (including Washington, DC and a value for all states) and each state's rate of participation for the year 2001.

##### 2. Does the data look complete? Are there any obvious issues with the observations?

The data does look complete.  The only issue I can really see is the large difference in participation in some states compared to others.  This could give misleading results in certain analyses.

##### 3. Create a data dictionary for the dataset.

```python
# Importing the data, creating a data dictionary, and printing the data
import csv
with open('./../assets/sat_scores.csv') as data_1:
    reader = csv.reader(data_1)
    headers = reader.next()
    data_dict
    for h in headers:
        data_dict[h] = []
    for row in reader:
        for j, k in zip(headers, row):
            data_dict[j].append(k)
    print data_dict
```

## Step 2: Load the data.

##### 4. Load the data into a list of lists

```python
# I did this with pandas first, but commented it out and continued with csv reader
#import pandas as pd

#data = pd.read_csv('./../assets/sat_scores.csv')
#data

# Loading data again, but this time into a list of lists named data_1_ls
# Also printing the data again
with open('./../assets/sat_scores.csv') as data_1:
    data_1_ls = []
    reader = csv.reader(data_1)
    for row in reader:
        data_1_ls.append(row)
print data_1_ls
```

##### 5. Print the data

```python
# Printing the data

print data_1_ls
```

##### 6. Extract a list of the labels from the data, and remove them from the data.

```python
# Extracted the column names from the data and assigned them as a list to labels with .pop()
labels = data_1_ls.pop(0)
print 'labels:', labels
print 'data:', data_1_ls
```

##### 7. Create a list of State names extracted from the data. (Hint: use the list of labels to index on the State column)

```python
# Created a list of state names named states_1

states_1 = [i[0] for i in data_1_ls]
print states_1
```

##### 8. Print the types of each column

```python
# Printed the types for the data of each column

print type(states_1[0])
print type([i[1] for i in data_1_ls][0])
print type([i[2] for i in data_1_ls][0])
print type([i[3] for i in data_1_ls][0])
```

##### 9. Do any types need to be reassigned? If so, go ahead and do it.

```python
# They are all strings, so Math, Verbal, and Rank need to be converted to floats in order to perform calculations
# on them.

# Changing the type to float for Math, Verbal, and Rank and storing it into new list of lists named r_data
# Also printing new list
r_data = []
for i in data_1_ls:
    for j in i:
        if i.index(j) == 0:
            r_data.append([j])
        else:
            r_data[data_1_ls.index(i)].append(float(j))
print r_data
```

##### 10. Create a dictionary for each column mapping the State to its respective value for that column. 

```python
# Creating a dictionary named stt_data for each column mapping the state to its respective value and printing it

stt_data = {i[0]: i[1:] for i in r_data}
print stt_data
```

##### 11. Create a dictionary with the values for each of the numeric columns

```python
# Creating separate lists for each column, then combining them all except states into num_data.
# Creating dictionary with the values for each of the numeric columns named dict_col and printing it
sts = [i[0] for i in r_data]
rt = [i[1] for i in r_data]
vrbl = [i[2] for i in r_data]
mth = [i[3] for i in r_data]
all_data = [sts] + [rt] + [vrbl] + [mth]
num_data = [rt] + [vrbl] + [mth]

dict_col = {i: j for i, j in zip(labels[1:], num_data)}
print dict_col
```

## Step 3: Describe the data

##### 12. Print the min and max of each column

```python
# Creating function to print min and max and calling it for rate, math, and verbal
def prnt_mam(col):
    print col
    print 'Min:', min(dict_col[col])
    print 'Max:', max(dict_col[col]), '\n'
prnt_mam('Rate')
prnt_mam('Math')
prnt_mam('Verbal')
```

```python
# Decided to go ahead and look at the mean and median of the data

import numpy as np
from scipy import stats

# Creating function to print the descriptive stats instead of writing it out three times
def des_stts(col):
    print col, '\n', 'Mean:', np.mean(dict_col[col]) 
    print 'Median:', np.median(dict_col[col])
    print 'Mode:', stats.mode(dict_col[col])[0], '\n'

# Calling the above function to print out mean, median, and mode of the variables
des_stts('Rate')
des_stts('Math')
des_stts('Verbal')
```

##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
# Using numpy to check my function
import numpy as np
print 'NUMPY RESULTS'
print 'Standard Deviation of Rate:', np.std(dict_col['Rate'])
print 'Standard Deviation of Math:', np.std(dict_col['Math'])
print 'Standard of Verbal:', np.std(dict_col['Verbal']), '\n'

# Creating my function for standard deviation and calling it to print the results
std_formula = lambda y: (sum([(i - sum(dict_col[y])/len(dict_col[y]))**2 for i in dict_col[y]])/len(dict_col[y]))**0.5
print 'MY LIST COMPREHENSION FUNCTION RESULTS'
print 'Standard Deviation of Rate:', std_formula('Rate')
print 'Standard Deviation of Math:', std_formula('Math')
print 'Standard of Verbal:', std_formula("Verbal")
```

## Step 4: Visualize the data

##### 14. Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
# Importing matplotlib and pyplot
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline

# Plotting rate in a histogram to look at its distribution.  It does not appear to be normally distributed.
plt.title('Rate Distribution', fontsize=24)
plt.xlabel('Rate', fontsize=18)
plt.ylabel('Frequency', fontsize=18)
plt.hist(dict_col['Rate'])
plt.show()
```

##### 15. Plot the Math distribution

```python
# Plotting math scores in a histogram to look at its distribution.  It almost appears to be normally distributed, 
# but positively skewed.
plt.title('Math Score Distribution', fontsize=24)
plt.xlabel('Score', fontsize=18)
plt.ylabel('Frequency', fontsize=18)
plt.hist(dict_col['Math'])
plt.show()
```

##### 16. Plot the Verbal distribution

```python
# Plotting verbal scores in a histogram to look at its distribution.  It does not appear to be normally distributed.
plt.title('Verbal Score Distribution', fontsize=24)
plt.xlabel('Verbal Score', fontsize=18)
plt.ylabel('Frequency', fontsize=18)
plt.hist(dict_col['Verbal'])
plt.show()
```

##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
# Creating scatterplots of the data to look for relationships between variables

plt.title('Rate vs. Verbal Scores', fontsize=24)
plt.xlabel('Rate', fontsize=18)
plt.ylabel('Verbal', fontsize=18)
plt.scatter(dict_col['Rate'], dict_col['Verbal'], s=80, alpha=0.5)
plt.show()

plt.title('Rate vs. Math Scores', fontsize=24)
plt.xlabel('Rate', fontsize=18)
plt.ylabel('Math', fontsize=18)
plt.scatter(dict_col['Rate'], dict_col['Math'], s=80, alpha=0.5)
plt.show()

plt.title('Verbal vs. Math Scores', fontsize=24)
plt.xlabel('Verbal', fontsize=18)
plt.ylabel('Math', fontsize=18)
plt.scatter(dict_col['Verbal'], dict_col['Math'], s=80, alpha=0.5)
plt.show()

# Importing pandas to create a dataframe to use with seaborn and importing seaborn
import pandas as pd

df = pd.DataFrame(dict_col)

import seaborn as sns

# Creating jointplots with seaborn to show histograms and scatterplots together
sns.jointplot(x='Verbal', y='Math', data=df, kind='reg')
sns.jointplot(x='Rate', y='Verbal', data=df, kind='reg')
sns.jointplot(x='Rate', y='Math', data=df, kind='reg')
```

##### 20. Are there any interesting relationships to note?

```python
# There seems to be a strong relationship between Math and Verbal scores, where higher verbal scores are correlated
# with higher math scores.

# There is also a relationship between rate and math/verbal scores.  These weren't as strong, but as rate went up, 
# math and verbal scores went down.
```

##### 21. Create box plots for each variable.
